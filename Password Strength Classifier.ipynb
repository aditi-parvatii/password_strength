```python
#converting the data given into array
password_tuple=np.array(data)
password_tuple
```

```python
#importing the random module
import random 
random.shuffle(password_tuple)
```

```python
x=[labels[0] for labels in password_tuple] 
y=[labels[1] for labels in password_tuple] 
```

```python
#this is the independent feature
x
```

```python
#this is the dependent feature
y
```

```python
def word_divide_char(input):
    character=[]
    for i in input:
        character.append(i)
    return character
```

```python
#getting the password character by character by using the function given above
word_divide_char('kzde5577')
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
#Converting a collection of raw documents to a matrix of TF-IDF features
vectorizer=TfidfVectorizer(tokenizer=word_divide_char)
```

```python
X=vectorizer.fit_transform(x)
```

```python
#The shape of the matrix is displayed
X.shape
```

```python
vectorizer.get_feature_names()
```

```python
first_document_vector=X[0]
first_document_vector
```

```python
#Returns a dense matrix representation of this first_document_vector matrix
first_document_vector.T.todense()
```

```python
#sorting the TF-IDF values in descending order
df=pd.DataFrame(first_document_vector.T.todense(),index=vectorizer.get_feature_names(),columns=['TF-IDF'])
df.sort_values(by=['TF-IDF'],ascending=False)
```

```python
from sklearn.model_selection import train_test_split
```

```python
#Splitting our testing and training data. We initialize the test size to be 20% and the training size to be 80% of the data
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2)
```

```python
X_train.shape
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
clf=LogisticRegression(random_state=0,multi_class='multinomial')
```

```python
clf.fit(X_train,y_train)
```

```python
#The output indicates that the password "%@123abcd" is of average strength password ie.1
dt=np.array(['%@123abcd'])
pred=vectorizer.transform(dt)
clf.predict(pred)
```

```python
y_pred=clf.predict(X_test)
y_pred
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score
```

```python
#dialgnals are true predictions
#accuracy of 81%
cm=confusion_matrix(y_test,y_pred)
print(cm)
print(accuracy_score(y_test,y_pred))
```

```python
#getting the classification report
from sklearn.metrics import classification_report
print(classification_report(y_test,y_pred))
```
